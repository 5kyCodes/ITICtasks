```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
data1 = data.groupby('Y')['Id'].count()
```

```python
plt.xlabel("Quality Rating")
plt.ylabel("Quantity")
plt.title("Quantity per Quality Rating")
labels = ['HQ','LQ EDIT','LQ CLOSE']
plt.bar(labels, data1)
plt.show()
```
