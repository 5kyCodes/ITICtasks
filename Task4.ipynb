# Task4 Feature Extraction on Coursera MOOC dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
threads = pd.read_csv("course_threads.csv")
infomation = pd.read_csv("course_information.csv")
posts = pd.read_csv("course_posts.csv")
```

```python
threads.head()
```

```python
#getting the sum of the number of views 
views = threads.groupby('thread_id')['num_views'].sum().reset_index(name='views_sum')
```

### Number of Views

```python
views.head()
```

```python
#transforming votes column values to absolute values and squaring them
posts['votes_squared'] = posts['votes'].abs().pow(2)
```

```python
posts.head(10)
```

```python
#getting the total sum of the square of the abs value of votes per thread id
vote_rating = posts.groupby('thread_id')['votes_squared'].sum().reset_index(name='vote_measure')
```

### Vote Rating

```python
vote_rating.head()
```

```python
#finding the maximum value of votes per thread id
arg_max = posts.groupby('thread_id')['votes'].idxmax().reset_index(name='arg_max')
```
